## 머신러닝 시작하기

```python
import tensorflow as tf
```

```python
print(tf.__version__)
```

```python
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 뉴런 1개, 입력 [1]

dense = Dense(units=1, input_shape=[1])
```

```python
# Sequential 클래스로 네트워크 층을 정의

model = Sequential([dense])
```

```python
# 옵티마이저 = sgd (stochastic gradient descent, 확률적 경사 하강법)
# 손실함수 = mean squared error

model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# Epoch = 훈련 데이터를 모두 사용해 모델을 1회 훈련하는 것을 지칭

model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10.0]))
```

```python
print("신경망이 학습한 것: {}".format(dense.get_weights()))
```

```python
print("신경망이 학습한 것: {}".format(model.layers[0].get_weights()))
```
